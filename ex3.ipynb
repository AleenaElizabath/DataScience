```python

#Qu1
import matplotlib.pyplot as mat
import numpy as np
ar1=np.array([1,2,6,18])
ar2=np.array([3,10,12,20])
mat.plot(ar1,ar2,'o:r',marker='o',ms=10,mfc="green",mec="green")
```

```python
#Qu2
temp=np.array([12,14,16,18,20,22,24])
sales=np.array([100,200,250,400,300,450,500])
mat.plot(temp,sales,'o:b')
```

```python
#Qu3
mat.suptitle("Age vs BMI")
x,y=np.loadtxt("textfile.txt",delimiter=',')
mat.plot(x,y,color="grey")
mat.xlabel("Age")
mat.ylabel("BMI")
```

```python
#Qu4
ar1=np.array([1,5,9,4,11,22])
ar3=np.array([11,5,19,54,11,6])
ar2=np.array([81,24,33,44,25,46])
mat.plot(ar1,color="red")
mat.plot(ar2,color="green")
mat.plot(ar3,color="blue")
mat.legend("rgb")
```

```python
#qu5
ar1=np.array([1,5,9,4,11,6])
ar3=np.array([11,5,19,54,11,6])
ar2=np.array([81,24,33,44,25,46])
ar4=np.array([8,24,3,44,65,46])
mat.subplot(2,2,1)
mat.plot(ar1,color='red',marker='o')
mat.subplot(2,2,2)
mat.plot(ar2,marker='d')
mat.subplot(2,2,3)
mat.plot(ar3,color='green',marker='^')
mat.subplot(2,2,4)
mat.plot(ar4,color='violet',marker='s')
```
