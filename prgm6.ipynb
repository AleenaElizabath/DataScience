```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn import metrics
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dis=pd.read_csv("test.csv")
print(dis)
x=dis[['sore_throat','fever','swoolen_glands','congestion','headache']]
y=dis[['diagnosis']]
fname=['sore_throat','fever','swoolen_glands','congestion','headache']
tname=['Strep_throat','Allergy','Cold']
```

```python
clf=DecisionTreeClassifier(criterion='entropy')
clf=clf.fit(x,y)
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=fname, class_names=tname, filled=True, rounded=True)
plt.title("Decision Tree Classifier for Iris Dataset")
plt.show()
```

```python
newcase=[[0,0,1,0,0]]
y_new=clf.predict(newcase)
print(y_new)
plt.figure()
tree.plot_tree(clf, feature_names=fname, class_names=tname, filled=True, rounded=True)
```
